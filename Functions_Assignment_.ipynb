**Q4. What is the purpose of the `return` statement in a Python function?**

Ans4- In Python, the return statement serves two primary purposes within a function:

Exiting the function: When a return statement is encountered during the execution of a function, the function immediately terminates. Any code within the function that appears after the return statement will not be executed.

Returning a value to the caller: The return statement can optionally be followed by an expression, which represents the value that the function will send back to the part of the code that called it. This returned value can then be used in further calculations, assigned to variables, or printed, among other things. If no value is specified after return (i.e., just return), the function implicitly returns None.

list:[47,11,42,13]?

Ans11- The reduce function, often found in functional programming paradigms, iteratively applies a given function to the elements of a sequence, accumulating a single result. For a sum operation on the list [47, 11, 42, 13], the internal mechanism unfolds as follows:

Initialization: The reduce function takes a binary function (in this case, an addition function) and the list [47, 11, 42, 13]. If an initial value is not explicitly provided, the first element of the list (47) is used as the initial accumulator.

First Iteration: The addition function is applied to the initial accumulator (47) and the next element in the list (11).
47 + 11 = 58
The result (58) becomes the new accumulator.

Second Iteration: The addition function is applied to the current accumulator (58) and the next element in the list (42).
58 + 42 = 100
The result (100) becomes the new accumulator.

Third Iteration: The addition function is applied to the current accumulator (100) and the final element in the list (13).
100 + 13 = 113
The result (113) becomes the new accumulator.

Final Result: Since there are no more elements in the list, the reduce function returns the final accumulated value, which is 113.

```python
# Q4. Write a Python function that checks if a given number is prime or not from 1 to 200?
numbers = range(1, 201)

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

for number in numbers:
    if is_prime(number):
        print(f"{number} is a prime number")
```

```python
# Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration
```

```python
# Q7. Implement a generator function that reads a file line by line and yields each line as a string?
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            yield line.strip() # yield each line after removing leading/trailing whitespace
```

```python
# Example usage:
file_generator = read_file_line_by_line("my_file.txt")

# Iterate through the lines using the generator
for line in file_generator:
    print(line)
```

```python
# Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit?
lis = [10, 20, 30, 40, 50]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_list = list(map(celsius_to_fahrenheit, lis))
print(fahrenheit_list)
```
